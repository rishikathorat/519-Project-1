```python
import anthropic
```

```python
import pandas as pd

df = pd.read_json("hf://datasets/theblackcat102/joke_explaination/main.jsonl", lines=True)
```

```python
df.head()
```

```python

client = anthropic.Anthropic(
    # defaults to os.environ.get("ANTHROPIC_API_KEY")
    api_key="",
)
```

```python
def generate_output(text):
    message = client.messages.create(
        model="claude-3-opus-20240229",
        max_tokens=1000,
        temperature=0,
        system= f"""
        You are tasked with detecting and explaining words or phrases that have multiple meanings in a given text, and determining if these meanings are appropriate for a specified age group.
        This task is important for understanding wordplay in jokes, puns, and riddles.
        Here is the text to analyze:
        <text>
        {text}
        </text>

        Please follow these steps:
        1. Carefully read the provided text.
        2. Identify any words or phrases that are used with multiple meanings or could be interpreted in different ways within the context of the text. Ensure that the meanings are fairly different; if they are very similar, then don't include that word.
        3. For each identified word or phrase, explain the different meanings or interpretations.
        4. Provide your analysis in the following format:


        Word/Phrase: [Insert the word or phrase with multiple meanings]
        Meaning 1: [Explain the first meaning or interpretation]
        Meaning 2: [Explain the second meaning or interpretation]
        (Add more meanings if applicable)
        

        If no words or phrases with multiple meanings are found, state this clearly.

        Here are two examples of good outputs:
        Example 1: 
  
        Word/Phrase: trunk
        Meaning 1: The long, flexible nose-like appendage of an elephant
        Meaning 2: A large container or box used for storing items

        Example 2:
        Word/Phrase: No words with multiple meanings found
        Explanation: The given text does not contain any words or phrases that are used with multiple meanings or could be interpreted in different ways within this context.

        Please begin your analysis now.
        """,
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": text
                    }
                ]
            }
        ]
    )
    #print(message.content)
    return message.content[0].text
```

```python
df = df[0:200]
df
```

```python
df['predicted_exp'] = df['joke'].apply(generate_output)
```

```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
```

```python

# Load pre-trained BERT model and tokenizer from Hugging Face
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Function to get the BERT embeddings for a given sentence
def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
    # Get the CLS token embedding
    embeddings = outputs.last_hidden_state[:, 0, :]  # CLS token is at index 0
    return embeddings.numpy()

# Compute cosine similarity between two columns row-wise
cosine_similarities = []

for text1, text2 in zip(df['explaination'], df['predicted_exp']):
    # Get BERT embeddings for each sentence
    embedding_1 = get_bert_embedding(text1)
    embedding_2 = get_bert_embedding(text2)
    
    # Compute cosine similarity
    similarity = cosine_similarity(embedding_1, embedding_2)[0][0]
    cosine_similarities.append(similarity)

# Add cosine similarity to DataFrame
df['cosine_similarity'] = cosine_similarities

# Display the DataFrame with similarities
print(df)
```
